```python
import pandas as pd
import numpy as np
df = pd.read_csv("Reviews_clean_chunks_emb.csv")
```

```python
df_val = df.sample(n=5, random_state=1)
row = df_val.iloc[1]

query_text = row["chunk_text"]
query_emb  = row["embedding"]
query_emb = np.fromstring(
    query_emb.strip("[]"),
    sep=" ",
    dtype=np.float32
)
```

I leave 5 rows for validation. And I use the second row here

```python
import chromadb
client = chromadb.PersistentClient(path="./chroma_db")
collection = client.get_or_create_collection(name="amazon_reviews_chunks")
results = collection.query(
    query_embeddings=[query_emb],
    n_results=5,
)
```
